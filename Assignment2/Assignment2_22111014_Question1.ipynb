```python
# importing all the necessary modules
import numpy as np
import conllu
import collections
import heapq
from collections import Counter
```

```python
# Reading conllu files with the help of conllu module
with open('hi_hdtb-ud-dev.conllu', mode="r", encoding="utf-8") as data:
    annotations = data.read()
sentences1 = conllu.parse(annotations)

with open('hi_hdtb-ud-train.conllu', mode="r", encoding="utf-8") as data:
    annotations = data.read()
sentences2 = conllu.parse(annotations)

with open('hi_hdtb-ud-test.conllu', mode="r", encoding="utf-8") as data:
    annotations = data.read()
sentences3 = conllu.parse(annotations)
```

```python
sentences = sentences1 + sentences2 + sentences3    # Merging all three files
print(len(sentences))
```

```python
sentences[0][0]
```

```python
# a
from collections import defaultdict
uni_pos_tags = defaultdict(int)     #upos 
detail_pos_tags = defaultdict(int)  #xpos
for _ in sentences:
    for i in range(len(_)):
        uni_pos_tags[_[i]['upos']] += 1
        detail_pos_tags[_[i]['xpos']] += 1
uni_pos_tags
```

```python
freq_uni_pos_tags = heapq.nlargest(len(uni_pos_tags), uni_pos_tags, key=uni_pos_tags.get) 
for i in freq_uni_pos_tags:
    print(i,uni_pos_tags[i])
```

```python
word = defaultdict(list)
word_freq = defaultdict(int)
for _ in sentences:
    for i in range(len(_)):
        word_freq[_[i]['form']] += 1
        word[_[i]['form']].append(_[i]['upos'])
```

```python
l = heapq.nlargest(len(word_freq), word_freq, key = word_freq.get)
for i in l:
    print(i,Counter(word[i]).most_common())
```

```python
# b
Each_pos_tags = defaultdict(list)
for _ in sentences:
    for i in range(len(_)):
        Each_pos_tags[_[i]['upos']].append(_[i]['form'])
```

```python
from collections import Counter
freq_words_each_pos_tag = defaultdict()
for i in Each_pos_tags.keys():
    temp_dict = Counter(Each_pos_tags[i])
    freq_words_each_pos_tag[i] = heapq.nlargest(50, temp_dict, key = temp_dict.get)
```

```python
freq_words_each_pos_tag # It is a dict which has different pos tags as a key and most freq 50 words corresponding to that pos tag as a value
```

```python
# cross-check 
Counter(Each_pos_tags['NOUN']).most_common()
```

```python
# c
Gender = defaultdict(int)
Case = defaultdict(int)
Number = defaultdict(int)
for _ in sentences:
    for i in range(len(_)):
        if not( _[i]['feats']):
            continue
        if 'Gender' in _[i]['feats'].keys():   # checking whether Gender present in the Feats or not.
            Gender[_[i]['feats']['Gender']] += 1
        if 'Case' in _[i]['feats'].keys():
            Case[_[i]['feats']['Case']] += 1
        if 'Number' in _[i]['feats'].keys():
            Number[_[i]['feats']['Number']] += 1
```

```python
print("Gender->")
freq_Gender = heapq.nlargest(len(Gender), Gender, key=Gender.get)
for i in freq_Gender:
    print(i,Gender[i])
print("Case->")
freq_Case = heapq.nlargest(len(Case), Case, key=Case.get)
for i in freq_Case:
    print(i,Case[i])
print("Number->")
freq_Number = heapq.nlargest(len(Number), Number, key=Number.get)
for i in freq_Number:
    print(i,Number[i])
```

```python
# d
Triple = defaultdict(int)
for _ in sentences:
    for i in range(len(_)):
        if not( _[i]['feats']):
            continue
        if 'Gender' in _[i]['feats'].keys() and 'Case' in _[i]['feats'].keys() and 'Number' in _[i]['feats'].keys():  # Tuple format ->(gender,case,number)
            Triple[(_[i]['feats']['Gender'],_[i]['feats']['Case'],_[i]['feats']['Number'])] += 1
Triple
```

```python
freq_Triple = heapq.nlargest(len(Triple), Triple, key=Triple.get)   # Tuples in decreasing order of their freq.
for i in freq_Triple:
    print(i,Triple[i])
```

```python
# e
Head_uni_pos_tags = defaultdict(int)
Head_detail_pos_tags = defaultdict(int)
for _ in sentences:
    for i in range(len(_)):
        if _[i]['misc']['ChunkType'] == 'head':
            Head_uni_pos_tags[_[i]['upos']] += 1
            Head_detail_pos_tags[_[i]['xpos']] += 1
Head_uni_pos_tags
```

```python
freq_head_tags = heapq.nlargest(len(Head_uni_pos_tags), Head_uni_pos_tags, key=Head_uni_pos_tags.get)
for i in freq_head_tags:
    print(i,Head_uni_pos_tags[i])
```

```python
Head_word = defaultdict(list)
Head = defaultdict(int)
for _ in sentences:
    for i in range(len(_)):
        if _[i]['misc']['ChunkType'] == 'head':
            Head[_[i]['form']] += 1
            Head_word[_[i]['form']].append(_[i]['upos'])
```

```python
l = heapq.nlargest(len(Head), Head, key=Head.get)
for i in l:
    print(i,Counter(Head_word[i]).most_common())
```

```python
# f and G 
Relation_total = defaultdict(int)
Relation_per_tuple = defaultdict(list)
Two_tuples_total = defaultdict(int)
Two_tuples_per_relation = defaultdict(list)
for _ in sentences:
    for i in range(len(_)):
        Relation_total[_[i]['deprel']] += 1
        if _[i]['head'] == 0:  # Root Word
            Relation_per_tuple[_[i]['deprel']].append((_[i]['upos'],'_'))
            Two_tuples_total[(_[i]['upos'],'_')] += 1
            Two_tuples_per_relation[(_[i]['upos'],'_')].append(_[i]['deprel'])
            continue
        Relation_per_tuple[_[i]['deprel']].append((_[i]['upos'],_[_[i]['head']-1]['upos']))
        Two_tuples_total[(_[i]['upos'],_[_[i]['head']-1]['upos'])] += 1
        Two_tuples_per_relation[(_[i]['upos'],_[_[i]['head']-1]['upos'])].append(_[i]['deprel'])
```

```python
# frequency of two tuple as total
freq_Two_tuple_total = heapq.nlargest(len(Two_tuples_total), Two_tuples_total, key=Two_tuples_total.get)
for i in freq_Two_tuple_total:  
    print(i,"->",Two_tuples_total[i])
```

```python
# freq of each two tuples with respect to each of their relation
for i in Two_tuples_per_relation.keys():
    Two_tuples_per_relation[i] = Counter(Two_tuples_per_relation[i]).most_common()
for i in freq_Two_tuple_total:
    print(i,"->",dict(Two_tuples_per_relation[i]))
```

```python
# freq of relation as total
freq_Relation_total = heapq.nlargest(len(Relation_total), Relation_total, key=Relation_total.get)
for i in freq_Relation_total:
    print(i,"->",Relation_total[i])
```

```python
# freq of relation with respect to each associated two tuple.
for i in Relation_per_tuple.keys():
    Relation_per_tuple[i] = Counter(Relation_per_tuple[i]).most_common()
for i in freq_Relation_total:
    print(i,"->",dict(Relation_per_tuple[i]))
```
